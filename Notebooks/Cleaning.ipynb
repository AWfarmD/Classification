```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, date
```

```python
df = pd.read_csv('Animal_Shelter_Intake_and_Outcome.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['Impound Number', 'Animal ID', 'Intake Date', 'Intake Subtype', 
         'Outcome Subtype', 'Intake Condition', 'Intake Jurisdiction', 
         'Outcome Jurisdiction', 'Outcome Zip Code', 'Location', 'Count'],
        axis=1, inplace=True)
```

```python
df.rename(columns = {'Name':'name',
                     'Type':'type',
                     'Breed':'breed',
                     'Color':'color',
                     'Sex':'sex',
                     'Size':'size',
                     'Date Of Birth':'dob',
                     'Kennel Number':'shelter',
                     'Outcome Date': 'outcome_date',
                     'Days in Shelter':'days_in_shelter',
                     'Intake Type':'intake_type',
                     'Outcome Type':'outcome',
                     'Outcome Condition':'outcome_condition'}, inplace=True)
```

```python
df.info()
```

```python
df.dropna(subset=['size', 'dob', 'outcome_date', 'outcome', 'outcome_condition'], inplace=True)
```

```python
df.name = df.name.fillna('no')
```

```python
df.name = df.name.apply(lambda x: 'yes' if x != 'no' else x)
```

```python
df.info()
```

```python
df.head()
```

```python
df.dob = pd.to_datetime(df.dob)
```

```python
df.outcome_date = pd.to_datetime(df.outcome_date)
```

```python
df['age'] = df.outcome_date - df.dob
```

```python
df.drop(['dob'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.age = df.age.astype(str)
```

```python
df.age = df.age.str.strip('days')
```

```python
df.age = df.age.astype(int)
```

```python
df.rename(columns = {'age':'age_in_days'}, inplace=True)
```

```python
df.sex.unique()
```

```python
drop_sex = df[df.sex == 'Unknown'].index
```

```python
df.drop(drop_sex, inplace=True)
```

```python
df.sex.unique()
```

```python
df.head()
```

```python
df['outcome_day_of_week'] = df.outcome_date.dt.dayofweek
```

```python
df.drop('outcome_date', axis=1, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('adopt.csv', index=False)
```
